```python
import numpy.random as rand
import matplotlib.pyplot as plt
import numpy as numpy
%matplotlib inline
```

```python
with open("test.txt") as file:
    x = [int(line.split(' ')[0]) for line in file]

with open("test.txt") as file:
    times = [float(line.split(' ')[2][:-1]) for line in file]

y = [times[0] / time for time in times]
```

```python
figure = plt.figure(figsize=(20, 10))
axis = figure.gca()

axis.plot(x, y, label="S(p)")
axis.legend(loc="upper left")
axis.set_xlabel("p", fontsize=20)
axis.set_ylabel(r"S", fontsize=20)
axis.set_title("S(p)")
```
